### 조사 Pathway에 속한 orthology 확인

```python
down_path = [
    'ko00472', 'ko05231', 'ko00430', 'ko00984', 'ko04150'   ,
    'ko03320', 'ko00333', 'ko00471', 'ko04750', 'ko00830',
    'ko00944', 'ko00404', 'ko04330', 'ko00565', 'ko04975'
]

up_path = [
    'ko04024', 'ko04977', 'ko00601', 'ko04530', 'ko04151',
    'ko04141', 'ko01053', 'ko00220', 'ko04120'
]
```

```python
import pandas as pd

# Up-regulated DataFrame에 'Regulation' 열 추가
significant_up_ortho_df = data[
    (data['FC_value'] > 2) & 
    (data['LogReg_p_fdr'] <= 0.05) & 
    (data['Wilcoxon_p'] <= 0.05)
]
significant_up_ortho_df = significant_up_ortho_df.reset_index(drop=True)
significant_up_ortho_df['Regulation'] = 'up-regulated'

# Down-regulated DataFrame에 'Regulation' 열 추가
significant_down_ortho_df = data[
    (data['FC_value'] < 0.5) & 
    (data['LogReg_p_fdr'] <= 0.05) & 
    (data['Wilcoxon_p'] <= 0.05)
]
significant_down_ortho_df = significant_down_ortho_df.reset_index(drop=True)
significant_down_ortho_df['Regulation'] = 'down-regulated'

# 두 DataFrame 합치기
significant_ortho_df = pd.concat([significant_up_ortho_df, significant_down_ortho_df], ignore_index=True)

# 결과 확인
significant_ortho_df
```

```python
pathway = up_path+down_path
pathway_df = pd.DataFrame({"KEGG_no": pathway})
path_label = pd.read_csv(file_path_input+"/pathway_label_list.csv")
pathway_df = pd.merge(pathway_df,path_label,how = 'left',left_on = 'KEGG_no', right_on = 'KEGG_no')
pathway_df
```

```python
ortho_with_path = pd.merge(pathway_df,significant_ortho_df, how = 'right',left_on = 'KEGG_no', right_on = "Pathway_ID").drop('KEGG_no',axis=1)
```
